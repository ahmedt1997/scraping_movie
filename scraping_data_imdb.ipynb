```python
# relevant packages & modules
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
```

```python
# initialize Chrome
driver = webdriver.Chrome('C:\webdriver\chromedriver.exe')
```

```python
driver.get('https://www.imdb.com/search/title/?title_type=feature&genres=crime&sort=num_votes,desc&ref_=adv_prv')
```

```python
#title
title = driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a')
```

```python
type(title)
```

```python
title[0].text
```

```python
title[10].text
```

```python
title_movie = []
for i in range(len(title)):
    title_movie.append(title[i].text)
```

```python
title_movie
```

```python
date = driver.find_elements_by_xpath('//span[@class="lister-item-year text-muted unbold"]')
```

```python
date[30].text
```

```python
date_movie = []
for i in range(len(date)):
    date_movie.append(date[i].text)
```

```python
#runtime
time = driver.find_elements_by_xpath('//span[@class="runtime"]')
```

```python
time_movie=[]
for i in range(len(time)):
    time_movie.append(time[i].text)
```

```python
# rating 
rating = driver.find_elements_by_xpath('//div[@class="inline-block ratings-imdb-rating"]/strong')
```

```python
rating_movie = []
for i in range(len(rating)):
    rating_movie.append(rating[i].text)
```

```python
# genre
```

```python
genre =  driver.find_elements_by_xpath('//span[@class="genre"]')
```

```python
genre_movie = []
for i in range(len(genre)):
    genre_movie.append(genre[i].text)
```

```python
genre_movie
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({"title":title_movie,"date":date_movie,"time":time_movie,"rating":rating_movie,"genre":genre_movie})
```

```python
df
```

```python
df.to_excel('dataset_crime_movies_imdb.xlsx',index=False)
```
